```python
from synthetic_generator import SyntheticDataGenerator
import pandas as pd
from datetime import datetime, timedelta
import random
```

create data

```python
subject = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
verbs = ['SEND', 'RECEIVE', 'TRANSFER', 'WITHDRAW', 'DEPOSIT']
objects = ['BTC', 'ETH', 'USDT']
```

```python
def generate_transactions(count=10):
    data = []

    current_date = datetime.now()

    for _ in range(count):
        transaction = {
            'subject': random.choice(subject),
            'verb': random.choice(verbs),
            'object': random.choice(objects + [f'WALLET_{random.randint(1000,9999)}']),
            #'amount': round(random.uniform(0.1, 10.0), 8),
            #'timestamp': current_date - timedelta(minutes=random.randint(0, 100))
        }
        data.append(transaction)

    df = pd.DataFrame(data)
    #df['timestamp'] = pd.to_datetime(df['timestamp'])

    return df#.sort_values('timestamp')
```

```python
df = generate_transactions(150)
```
